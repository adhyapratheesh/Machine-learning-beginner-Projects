```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data=pd.read_csv("/content/movie.csv")
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
selected_features=['genres','keywords','tagline','overview','cast','director']
```

```python
for feature in selected_features:
  movie_data[feature]=movie_data[feature].fillna('')
```

```python
combine_features=movie_data['genres']+" "+movie_data['keywords']+" "+movie_data['tagline']+" "+movie_data['overview']+" "+movie_data['cast']+" "+movie_data['director']
```

```python
print(combine_features)
```

```python
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combine_features)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
movie_name=input("Enter the movie name:")
```

```python
all_titles=movie_data['title'].tolist()
```

```python
find_close_one=difflib.get_close_matches(movie_name,all_titles)
print(find_close_one)
```

```python
close_match=find_close_one[0]
print(close_match)
```

```python
index_movie=movie_data[movie_data.title==close_match]['index'].values[0]
print(index_movie)
```

```python
similarity_score=list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
sorted_movie_list=sorted(similarity_score,key = lambda x :x[1],reverse=True)
print(sorted_movie_list)
```

```python
print("Movie suggestions are :")
i=1;
for movie in sorted_movie_list[1:]:
  index=movie[0]
  movie_names=movie_data[movie_data.index==index]["title"].values[0]
  if i<20:
    print(i,".",movie_names)
    i=i+1

```

MOVIE RECOMMENDATION SYSYTEM

```python
movie_name=input("Enter the movie name :")
all_titles=movie_data['title'].tolist()
find_close_one=difflib.get_close_matches(movie_name,all_titles)
close_match=find_close_one[0]
index_movie=movie_data[movie_data.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_movie]))
sorted_movie_list=sorted(similarity_score,key = lambda x :x[1],reverse=True)
print("Movie suggestions are :")
i=1;
for movie in sorted_movie_list[1:]:
  index=movie[0]
  movie_names=movie_data[movie_data.index==index]["title"].values[0]
  if i<=20:
    print(i,".",movie_names)
    i=i+1
```
